## Desafío 3

### Condiciones array2

```python
for i in array2:
    if i%2==0 or i%3==0:
        print(str(i)+ " es divisible por 2 o 3")
    if i%2==0 and i%3==0:
        print(str(i)+ " es divisible por 2 y 3")
    if i%2!=0 and i%3==0:
        print(str(i)+ " es divisible por 3 y no por 2")
    if i%2!=0 and i%3!=0:
        print(str(i)+ " no es divisible por 2 ni 3")
    print("\n")
```

## Desafío 6

```python
desviacion=np.std(archivo["passengers"])
print(desviacion)
```

```python
archivo["outlier"]=0
```

```python
archivo
```

```python
archivo.at[archivo["passengers"] < media-desviacion, "outlier"]=1
archivo.at[archivo["passengers"] > media+desviacion, "outlier"]=1
```
